```python
urls = [
    "https://rss.app/feeds/6AmVlQlK9dsBw8wN.xml",
    "https://rss.app/feeds/KvgHC9qLF6n9wjUl.xml",
    "https://rss.app/feeds/hnKYtEFIUu6Chp8d.xml",
    "https://rss.app/feeds/1cJ4rTKNndABIeAw.xml",
    "https://rss.app/feeds/MIlJ7OfI8bH8XtEB.xml",
    "https://rss.app/feeds/WbHDL4YmIeAYAgT6.xml",
]

feeds = []
for _ in urls:
    feed = feedparser.parse(_)
    feeds.append(feed.entries)

feeds
```

```python
data = []

for _ in feeds:
    for i in _:
        summary = i.summary
        soup = BeautifulSoup(summary, "html.parser")

        datum = {
            "title": i.title.lower(),
            "summary": soup.div.get_text().lower(),
        }

        data.append(datum)
```

```python
# TODO: DataFrame
import pandas as pd

results = pd.DataFrame(data)
results
```

```python
# TODO: Ranked List
# TODO: Wiki Re-direction

import wikipedia

wiki_resources = []
for _ in results:
    try:
        wiki_resources.append(wikipedia.search(_))

    except:
        wiki_resources.append(None)
```
